# Iris data basic plots

## Base data setup

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

import numpy as np
import pandas as pd

iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
iris_df['target'] = iris_df['target'].map({0:"setosa", 1:"versicolor", 2:"virginica"})

X_iris_df = iris_df.iloc[:,:-1]
y_iris_df = iris_df.iloc[:,[-1]]

setosa=iris_df[iris_df['target']=='setosa']
versicolor =iris_df[iris_df['target']=='versicolor']
virginica =iris_df[iris_df['target']=='virginica']
```

### Scatter

```python
# https://www.kaggle.com/abhishekkrg/python-iris-data-visualization-and-explanation
#Plotting Petal Length vs Petal Width & Sepal Length vs Sepal width
#warnings.simplefilter("ignore")#Supress any warning
import matplotlib.pyplot as plt

plt.figure()
fig,ax=plt.subplots(1,2,figsize=(17, 9))
iris_df.plot(x="sepal length (cm)",y="sepal width (cm)",kind="scatter",ax=ax[0],sharex=False,sharey=False,label="sepal",color='r')
iris_df.plot(x="petal length (cm)",y="petal width (cm)",kind="scatter",ax=ax[1],sharex=False,sharey=False,label="petal",color='b')
ax[0].set(title='Sepal comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()
# plt.show()
# plt.close()

# we can see that  there are some petals which are smaller than rest of petal.
#Let's examine them
```

adding colour separation

```python
# https://www.kaggle.com/abhishekkrg/python-iris-data-visualization-and-explanation
#for each Species ,let's check what is petal and sepal distibutuon
plt.figure()
fig,ax=plt.subplots(1,2,figsize=(21, 10))

setosa.plot(x="sepal length (cm)", y="sepal width (cm)", kind="scatter",ax=ax[0],label='setosa',color='r')
versicolor.plot(x="sepal length (cm)",y="sepal width (cm)",kind="scatter",ax=ax[0],label='versicolor',color='b')
virginica.plot(x="sepal length (cm)", y="sepal width (cm)", kind="scatter", ax=ax[0], label='virginica', color='g')

setosa.plot(x="petal length (cm)", y="petal width (cm)", kind="scatter",ax=ax[1],label='setosa',color='r')
versicolor.plot(x="petal length (cm)",y="petal width (cm)",kind="scatter",ax=ax[1],label='versicolor',color='b')
virginica.plot(x="petal length (cm)", y="petal width (cm)", kind="scatter", ax=ax[1], label='virginica', color='g')

ax[0].set(title='Sepal comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()
# plt.show()
# plt.close()

#satosa   - satosa Petal are relatively smaller than rest of species .can be easily separable from rest of Species 
#versicolor & virginica are also separable in Petal comprasion
#satoa sepal are smallest in length and largest in Width than other species
```

### Histogram

```python
###############################Histogram# Plot all feature for all species###########################################################################
# from https://www.kaggle.com/abhishekkrg/python-iris-data-visualization-and-explanation 
plt.figure()

fig,ax=plt.subplots(4,3,figsize=(17, 8))
setosa["sepal length (cm)"].plot(kind="hist", ax=ax[0][0],label="setosa",color ='r',fontsize=10)
versicolor["sepal length (cm)"].plot(kind="hist", ax=ax[0][1],label="versicolor",color='b',fontsize=10)
virginica["sepal length (cm)"].plot( kind="hist",ax=ax[0][2],label="virginica",color='g',fontsize=10)

setosa["sepal width (cm)"].plot(kind="hist", ax=ax[1][0],label="setosa",color ='r',fontsize=10)
versicolor["sepal width (cm)"].plot(kind="hist", ax=ax[1][1],label="versicolor",color='b',fontsize=10)
virginica["sepal width (cm)"].plot( kind="hist",ax=ax[1][2],label="virginica",color='g',fontsize=10)

setosa["petal length (cm)"].plot(kind="hist", ax=ax[2][0],label="setosa",color ='r',fontsize=10)
versicolor["petal length (cm)"].plot(kind="hist", ax=ax[2][1],label="versicolor",color='b',fontsize=10)
virginica["petal length (cm)"].plot( kind="hist",ax=ax[2][2],label="virginica",color='g',fontsize=10)


setosa["petal width (cm)"].plot(kind="hist", ax=ax[3][0],label="setosa",color ='r',fontsize=10)
versicolor["petal width (cm)"].plot(kind="hist", ax=ax[3][1],label="versicolor",color='b',fontsize=10)
virginica["petal width (cm)"].plot( kind="hist",ax=ax[3][2],label="virginica",color='g',fontsize=10)

plt.rcParams.update({'font.size': 10})
plt.tight_layout()

ax[0][0].set(title='sepal length (cm)')
ax[0][1].set(title='sepal length (cm)')
ax[0][2].set(title='sepal length (cm)')
ax[1][0].set(title='sepal width (cm) ')
ax[1][1].set(title='sepal width (cm) ')
ax[1][2].set(title='sepal width (cm) ')
ax[2][0].set(title='petal length (cm)')
ax[2][1].set(title='petal length (cm) ')
ax[2][2].set(title='petal length (cm)')
ax[3][0].set(title='petal width (cm)')
ax[3][1].set(title='petal width (cm)')
ax[3][2].set(title='petal width (cm)')

ax[0][0].legend()
ax[0][1].legend()
ax[0][2].legend()
ax[1][0].legend()
ax[1][1].legend()
ax[1][2].legend()
ax[2][0].legend()
ax[2][1].legend()
ax[2][2].legend()
ax[3][0].legend()
ax[3][1].legend()
ax[3][2].legend()


plt.show()
plt.close()

###########################################################################################################
```
