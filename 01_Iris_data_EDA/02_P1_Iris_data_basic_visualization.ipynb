## Base data setup

```python
# Loading dataset from Scikit-Learn 0.22.2.post1 (Google Colab)
# From Kcikit-Learn 0.23, direct Data Frame import is possible with "as_frame=True"
# However, the day this notebook was compiled Google Colab used 0.22
# Hence the data trasformation with numpy and pandas libraries

# Data import from Scikit-Learn: Data type: Bunch (ndarray)
from sklearn.datasets import load_iris

# Separating Iris data set(Seoal/Petal) and target result (Iris Type)
# Data Type Bunch (ndarray)
# Prepairing the data for Plotting
# Sepal/Petal Length/Width = X, Iris Type = y
# NOTE: iris (Data Type Bunch (ndarray)) will be used for Data Frame transformation
iris = load_iris()
X = iris.data
y = iris.target

# Importing numpy and pandas for data transform
import numpy as np
import pandas as pd

# Creating pandas Data Fram with Iris (Data Type Bunch (ndarray)) above
# Combining the two data and target arraies into iris_df
# Column labels with feature names and target
# replace 0, 1, 2 with appropriate iris type
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
iris_df['target'] = iris_df['target'].map({0:"setosa", 1:"versicolor", 2:"virginica"})

# Prepairing the data frame for Plotting
# Sepal/Petal Length/Width = X, Iris Type = y
X_iris_df = iris_df.iloc[:,:-1]
y_iris_df = iris_df.iloc[:,[-1]]

# Prepairing the data frame for Plotting
# Sepaarting data based of its Iris type
setosa=iris_df[iris_df['target']=='setosa']
versicolor =iris_df[iris_df['target']=='versicolor']
virginica =iris_df[iris_df['target']=='virginica']

#
target_df = [iris_df[iris_df['target']=="setosa"]['sepal length (cm)'], 
             iris_df[iris_df['target']=="versicolor"]['sepal length (cm)'],
             iris_df[iris_df['target']=="virginica"]['sepal length (cm)']]
```

## Plots

### Scatter

```python
# https://www.kaggle.com/abhishekkrg/python-iris-data-visualization-and-explanation
#Plotting Petal Length vs Petal Width & Sepal Length vs Sepal width
#warnings.simplefilter("ignore")#Supress any warning
import matplotlib.pyplot as plt

plt.figure()
fig,ax=plt.subplots(1,2,figsize=(17, 9))
iris_df.plot(x="sepal length (cm)",y="sepal width (cm)",kind="scatter",ax=ax[0],sharex=False,sharey=False,label="sepal",color='r')
iris_df.plot(x="petal length (cm)",y="petal width (cm)",kind="scatter",ax=ax[1],sharex=False,sharey=False,label="petal",color='b')
ax[0].set(title='Sepal comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()
# plt.show()
# plt.close()

# we can see that  there are some petals which are smaller than rest of petal.
#Let's examine them
```

## KNN setup

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

### KNN plot and modelling

```python
# experimenting with different n values
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    y_pred = knn.predict(X)
    scores.append(metrics.accuracy_score(y, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```

```python
logreg = LogisticRegression()
logreg.fit(X, y)
y_pred = logreg.predict(X)
print(metrics.accuracy_score(y, y_pred))
```

### Split dataset: training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# experimenting with different n values
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```
