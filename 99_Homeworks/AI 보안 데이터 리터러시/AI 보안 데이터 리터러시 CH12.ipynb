```python
a = {'name': ['hong', 'ppl1', 'ppl2'], 'age':[20,30,25]}
```

```python
df1 = pd.DataFrame(a)
```

```python
df1
```

```python
b = {'name': ['hong', 'ppl1', 'ppl2'], 'city':['seoul', 'deajeon', 'seoul']}
```

```python
df2 = pd.DataFrame(b)
```

```python
df2
```

```python
df = pd.merge(left=df1, 
              right = df2, 
              how = 'left', 
              left_on = 'name', 
              right_on ='name')
```

```python
df
```

```python
x1 = (df['city'] == 'seoul')
```

```python
df[x1]
```

```python
x2 = (df['city'] != 'seoul')
```

```python
df[x2]
```

```python
x3 = (df['age']<30)
```

```python
df[x3]
```

```python
x4 = (df['city']=='seoul') & (df['age']<30)
```

```python
df[x4]
```

```python
x5 = (df['city']=='seoul') | (df['age']<30)
```

```python
df[x5]
```
