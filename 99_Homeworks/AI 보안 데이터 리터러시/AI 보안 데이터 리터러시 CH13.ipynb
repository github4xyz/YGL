## matplotlib

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.groupby(iris['species']).mean()
```

```python
df = iris.groupby(iris['species']).mean()
df.plot()
```

```python
df.plot.bar()
```

```python
df2 = iris.groupby(iris['species']).median()
```

```python
df2.plot()
```

```python
df2.plot.bar(rot=0)
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width')
```

```python
x = np.where(iris['species'] == 'setosa', 'red', 'green')
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = x)
```

```python
x = np.where(iris['species'] == 'setosa', 'red', 
    np.where(iris['species'] == 'versicolor', 'green', 'blue'))
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = x)
```

```python
iris.plot.scatter(x = 'petal_length', y = 'petal_width', c = x)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.groupby(tips['time']).mean()
```

```python
tips.groupby(tips['day']).mean()
```

```python
df3 = tips.groupby(tips['day']).mean()
```

```python
df3.plot()
```

```python
df3.plot.bar(rot=0)
```

```python
tips.corr()
```

```python
tips.plot.scatter(x = 'total_bill', y = 'tip')
```

```python
x = np.where(tips['day'] == 'Thur', 'red',
    np.where(tips['day'] == 'Fri', 'orange',
    np.where(tips['day'] == 'Sat', 'green', 'blue')))
```

```python
tips.plot.scatter(x = 'total_bill', y = 'tip', c = x)
```
